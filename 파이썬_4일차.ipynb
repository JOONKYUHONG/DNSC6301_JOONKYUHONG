# 제어문 : 조건문, 반복문

- 조건문 : if -> 결과를 bool (부울) 함수로 판단 -> True / False 
- 반복문 : for, while 
    - for 구문을 사용하는 경우: 반복 횟수가 정해진 경우
    - while 구문을 사용하는 경우: 조건을 만족 할떄 까지

- 부울 함수를 사용할 떄 비교연산자를 함께 사용한다
|비교연산|설명|
|-|-|
|A < B |A가 B보다 작다|
|A > B |A가 B보다 크다|
|A <= B |A가 B보다 작거나 같다|
|A >= B |A가 B보다 크거나 같다|
|A == B |A가 B와 같다|
|A != B |A가 B와 같지 않다|

-  주의 : 비교연산을 진행할떄도 값의 자료형을 변환해주어야 한다

## (3) if ~ else, if ~elif ~ else 구문을 사용하는 경우

```python
import datetime

now= datetime.datetime.now()

if now.hour >= 12:
    print("현재 시간은 오후 {}시 {}분 입니다".format(now.hour,now.minute))

else:  # 그렇지 않으면
    print("현재 시간은 오전 {}시 {}분 입니다".format(now.hour,now.minute))
```

```python
## 이렇게 안씀 
## else 는 하나의 if만 처리함
import datetime

now= datetime.datetime.now()

if 3 <= now.month <= 5:  # 1번 if
    print('현재 계절은 봄입니다')
if 6 <= now.month <= 8:  # 2번 if
    print('현재 계절은 여름입니다')
if 9 <= now.month <= 11: # 3번 if
    print('현재 계절은 가을입니다')

else:  # 여기선 3번 조건만 else 하는 경우
  print('현재 계절은 겨울입니다')
```
